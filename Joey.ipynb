```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("courses.csv")
data.head()
```

```python
grades = data[['Course Title', 'A+', 'A', 'A-', 'B+', "B", 'B-', 'C+', 'C', 'C-', 'D+', 'D', 'D-', 'F', 'W']]

grades 
```

```python
courses = grades.groupby('Course Title')

courses
```

```python
student_count = data[('A')] + data[('A+')] + data[('A-')] + data[('B')] + data[('B+')] + data[('B-')] + data[('C')] + data[('C+')] + data[('C-')] + data[('D')] + data[('D+')] + data[('D-')] + data[('F')]

points = data[('A')] * 4.0 + data[('A+')] * 4.0 + data[('A-')] * 3.67 + data[('B')] * 3.0 + data[('B+')] * 3.33 + data[('B-')] * 2.67 + data[('C')] * 2.0 + data[('C+')] * 2.33 + data[('C-')] * 1.67 + data[('D')] * 1.0 + data[('D+')] * 1.33 + data[('D-')] * .67 

gpa = points / student_count

data["GPA"] = gpa

data
```

```python
gpa_course_alltime = data.groupby([data["Subject"], data["Course Title"]])['GPA'].mean().reset_index()
gpa_course_alltime.head()
```

graph comparing the gpa distributions of various courses taught by a professor

```python
def graph2(professor):
        temp = gpa_course_alltime[(gpa_course_alltime['Primary Instructor'] == professor)]
        # x-coordinates of left sides of bars 
        left = temp['Course Title']
        
        # heights of bars
        height = temp['GPA']

        # plotting a bar chart
        plt.bar(left, height,
                width = 0.5, color = ['orange', 'blue'])
        
        # naming the x-axis
        plt.xlabel(left)
        # naming the y-axis
        plt.ylabel('Count')
        # plot title
        plt.title('Grade')
        
        # function to show the plot
        return plt.show()
```

```python
graph2('Manfroi, Aldo J')
```
