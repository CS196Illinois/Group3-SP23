Calculate GPA

```python
import pandas as pd 
data = pd.read_csv("courses.csv")
data.head()
```

```python
count=data[('A')]+data[('A+')]+data[('A-')]+data[('B')]+data[('B+')]+data[('B-')]+data[('C')]+data[('C+')]+data[('C-')]+data[('D')]+data[('D+')]+data[('D-')]+data[('F')]
points = data[('A')]*4+data[('A+')]*4+data[('A-')]*3.67+data[('B')]*3+data[('B+')]*3.33+data[('B-')]*2.67+data[('C')]*2+data[('C+')]*2.33+data[('C-')]*1.67+data[('D')]*1+data[('D+')]*1.33+data[('D-')]*0.67+data[('F')]*0
GPA=points/count
GPA
```

Function that takes in a major and finds the professors with the top 5 average grades and the lowest 5 average grades

```python
major = 'CS'
majordata = data[data['Subject'] == major]


instructor = majordata.groupby('Primary Instructor').sum()[['A', 'A+', 'A-', 'B', 'B+', 'B-', 'C', 'C+', 'C-', 'D', 'D+', 'D-', 'F', 'W']]
instructor['count'] = instructor.sum(axis=1)
instructor['points'] = instructor['A+']*4 + instructor['A']*4 + instructor['A-']*3.67 + instructor['B+']*3.33 + instructor['B']*3 + instructor['B-']*2.67 + instructor['C+']*2.33 + instructor['C']*2 + instructor['C-']*1.67 + instructor['D+']*1.33 + instructor['D']*1 + instructor['D-']*0.67 + instructor['F']*0
instructor['GPA'] = instructor['points'] / instructor['count']


top_5 = instructor.sort_values(by='GPA', ascending=False).head(5)
bottom_5 = instructor.sort_values(by='GPA').head(5)

print(top_5, bottom_5)
```

```python
# Replace 'COURSE_CODE' with the course code you want to find the professor's average GPA for
#course = 'Intro Asian American Studies'
course = 'ASS100'
# Filter the data by course code
#coursedata = data[data['Course Title'] == course]

coursedata = data[(data['Subject'] == course[:3]) & (data['Number'] == int(course[-3:]))]

print("coursedata: ", coursedata)
# Group the data by primary instructor and calculate their GPA
instructor = coursedata.groupby('Primary Instructor').sum()[['A', 'A+', 'A-', 'B', 'B+', 'B-', 'C', 'C+', 'C-', 'D', 'D+', 'D-', 'F', 'W']]
instructor['count'] = instructor.sum(axis=1)
instructor['points'] = instructor['A+']*4 + instructor['A']*4 + instructor['A-']*3.67 + instructor['B+']*3.33 + instructor['B']*3 + instructor['B-']*2.67 + instructor['C+']*2.33 + instructor['C']*2 + instructor['C-']*1.67 + instructor['D+']*1.33 + instructor['D']*1 + instructor['D-']*0.67 + instructor['F']*0
instructor['GPA'] = instructor['points'] / instructor['count']

# Sort the data by GPA and show the top 5 and bottom 5 results
top_5 = instructor.sort_values(by='GPA', ascending=False).head(5)
bottom_5 = instructor.sort_values(by='GPA').head(5)

# Print the results
print(f"Professor's average GPA for course {course}: {instructor['GPA'].mean()}")
print("Top 5 instructors by GPA:\n", top_5)
print("Bottom 5 instructors by GPA:\n", bottom_5)
```
