# Group By
Groupby is a more difficult concept (but not that difficult!), but essentially it groups all the data in the dataset by a column and then apply some sort of function you choose (like the median, average, the count, etc). It makes it easier to focus our analysis on a certain aspect of our data. This function also returns a dataframe, so you can use all the dataframe methods on it (like mean, std, sort)

```python
# Here is an example of a group by, I want to group by Subjects and then apply a count function (count how many rows there were)
groupby_subject_count = data.groupby("Subject").agg("count").reset_index()
# At the end of each groupby, you must but a .reset_index() or the dataframe will looked unorganized
groupby_subject_count
```

So what did this do? It basically took every single course of the same type, counted how many rows there were, and replaced the number in the datasets with them. So for example, it took every single CS course offered at UIUC, counted how many there were, and placed that number in dataframe. Note that the subject row is now replaced with each unique course, so there is only one row that represents the CS category

```python
# Here is the CS row, basically means that there is 2066 classes/sections in CS
groupby_subject_count[groupby_subject_count['Subject'] == 'CS']
```

Here is a more intuitive example, what if I wanted to find the mean of all letter grades in each section using groupby?

```python
groupby_subject_mean = data.groupby("Subject").agg("mean").reset_index()
groupby_subject_mean
```

Now we have different data in each column

```python
# Once again, I will find the CS category
groupby_subject_mean[groupby_subject_mean['Subject'] == 'CS']
```

Here is an explanation of these values. For the column "A+", it is showing the average number of A+s in CS classes at UIUC, the same goes for all the other letter grades. Some of this data is "useless" or "meaningless". For example, there is not much use for the average year of the class.

```python
# Another example, this time by professor. get the mean of the grades by professors
professor_mean = data.groupby("Primary Instructor").agg("mean").reset_index()
professor_mean
```

```python
# Sorting these by A+s, we can see which professor gave the most A+s!
professor_mean.sort_values("A+", ascending=False)
```
